```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors 
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = "Times New Roman"
import pickle
import treecorr
from astropy.io import fits
import os
```

```python
for i in range(1000):
    # Load dictionaries from the pickled file
    with open(f'/Users/danie/Library/CloudStorage/OneDrive-Personal/Physics UCL/Year 3/Astro Research Project 2023/ellipticities_LHsim/LH{i}_ellipticities_extracted.pkl', 'rb') as f:
        
        data = pickle.load(f)
        pos_g = data['posg']
        ell = data['ellipticities']
        #pos_dm = data['dm_den']

    print(pos_g.shape)
    print(ell['x'].shape)
    # print(pos_dm.shape)

    ##############################

    # Create a FITS table
    data = fits.BinTableHDU.from_columns([
        #fits.Column(name='jackknife_ID', format='D', array=jack),
        fits.Column(name='x', format='D', array=pos_g[:,2]),
        fits.Column(name='y', format='D', array=pos_g[:,0]),
        fits.Column(name='z', format='D', array=pos_g[:,1]),
        fits.Column(name='ia_1', format='D', array=ell['y'].real),
        fits.Column(name='ia_2', format='D', array=ell['y'].imag),
    ])

    # Create a Primary HDU
    primary_hdu = fits.PrimaryHDU()

    # Create an HDU list
    hdul = fits.HDUList([primary_hdu, data])

    # # Save the FITS file
    hdul.writeto(f'LH_ellipticities/CAMELS_data_LHy{i}.fits', overwrite=True)

    ###########################

    length=len(pos_g[:,0])
    x_rand = np.random.uniform(min(pos_g[:,0]), max(pos_g[:,0]), length)
    y_rand = np.random.uniform(min(pos_g[:,1]), max(pos_g[:,1]), length)
    z_rand = np.random.uniform(min(pos_g[:,2]), max(pos_g[:,2]), length)
    # z_true_rand = np.random.uniform(min(z_true), max(z_true), length)
    #z_true_rand2 = np.random.uniform(min(z_true), max(z_true), length)
    ia_1_rand = np.random.normal(loc=0, scale=np.std(ell['x'].real), size=length)
    ia_2_rand = np.random.normal(loc=0, scale=np.std(ell['x'].imag), size=length)

    # Create a FITS table
    data = fits.BinTableHDU.from_columns([
        #fits.Column(name='jackknife_ID', format='D', array=jack),
        fits.Column(name='x', format='D', array=x_rand),
        fits.Column(name='y', format='D', array=y_rand),
        fits.Column(name='z', format='D', array=z_rand),#0.1001*np.ones(len(pos_g[:,0]))),
        fits.Column(name='ia_1', format='D', array=ia_1_rand),
        fits.Column(name='ia_2', format='D', array=ia_2_rand),
    ])

    # Create a Primary HDU
    primary_hdu = fits.PrimaryHDU()

    # Create an HDU list
    hdul = fits.HDUList([primary_hdu, data])

    # Save the FITS file
    hdul.writeto(f'LH_ellipticities/CAMELS_rand_LHy{i}.fits', overwrite=True)
```
